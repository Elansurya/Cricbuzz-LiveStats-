```python

import pandas as pd
import mysql.connector
import numpy as np

# --- SAMPLE DATA ---
data = [
    # [player, format, matches, runs, batting_avg]
    ["Virat Kohli",       "Test",  90,  7756, 52.31],
    ["Virat Kohli",       "ODI",   275, 13430, 58.52],
    ["Virat Kohli",       "T20I",  125, 4200, 49.10],

    ["Rohit Sharma",      "Test",  65,  5500, 44.05],
    ["Rohit Sharma",      "ODI",   250, 10801, 48.60],
    ["Rohit Sharma",      "T20I",  150, 4100, 35.20],

    ["Steve Smith",       "Test",  110, 13672, 61.81],
    ["Steve Smith",       "ODI",   120, 5872, 43.89],

    ["Babar Azam",        "Test",  55,  4200, 45.65],
    ["Babar Azam",        "ODI",   170, 9137, 54.83],
    ["Babar Azam",        "T20I",  130, 3820, 37.05],

    ["Joe Root",          "Test",  145, 13000, 51.20],
    ["Joe Root",          "ODI",   105, 6543, 44.10],

    ["Kane Williamson",   "Test",  100, 9500, 52.55],
    ["Kane Williamson",   "ODI",   160, 6789, 47.21],
    ["Kane Williamson",   "T20I",  80,  2200, 33.85],

    ["Pat Cummins",       "Test",  75,  900,  12.16],   # Only 1 format with meaningful batting

    ["David Warner",      "Test",  105, 9695, 44.40],
    ["David Warner",      "ODI",   140, 6234, 41.75],
    ["David Warner",      "T20I",  90,  2750, 34.38],
]

columns = ["player_name", "format", "matches", "runs", "batting_avg"]
df = pd.DataFrame(data, columns=columns)

# --- PANDAS PROCESSING ---
# Count how many formats each player has played
format_count = df.groupby("player_name")["format"].nunique().reset_index()
format_count.columns = ["player_name", "num_formats"]
multi_format_players = format_count[format_count["num_formats"] >= 2]["player_name"].tolist()

filtered = df[df["player_name"].isin(multi_format_players)]

# Pivot to show runs per format
pivot = filtered.pivot_table(index="player_name", columns="format", values="runs", aggfunc="sum")
pivot = pivot.reindex(columns=["Test", "ODI", "T20I"])
pivot = pivot.fillna(0).astype(int)

# Overall average
overall_avg = filtered.groupby("player_name").apply(
    lambda x: round((x["runs"].sum() / x["matches"].sum()) * (x["batting_avg"].mean() / (x["runs"].sum() / x["matches"].sum())), 2)
).reset_index()
overall_avg.columns = ["player_name", "overall_avg"]

result = pivot.reset_index().merge(overall_avg, on="player_name")
result = result.sort_values("overall_avg", ascending=False).reset_index(drop=True)
result.index += 1
result.index.name = "Rank"

print("=" * 80)
print("QUESTION 11: Player Performance Across Formats (2+ formats played)")
print("=" * 80)
print(result.rename(columns={
    "player_name": "Player", "overall_avg": "Overall Avg"
}).to_string())

# --- MySQL ---
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Elan@2004",
    database="cricbuzz_livestats"
)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS player_format_performance (
    id INT AUTO_INCREMENT PRIMARY KEY,
    player_name VARCHAR(100),
    format VARCHAR(10),
    matches INT,
    runs INT,
    batting_avg FLOAT
)
""")
cursor.execute("DELETE FROM player_format_performance")

for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO player_format_performance (player_name, format, matches, runs, batting_avg)
        VALUES (%s, %s, %s, %s, %s)
    """, (row["player_name"], row["format"], int(row["matches"]),
          int(row["runs"]), float(row["batting_avg"])))

conn.commit()

# --- SQL QUERY ---
print("\n--- SQL QUERY OUTPUT ---")
cursor.execute("""
    SELECT 
        p.player_name,
        MAX(CASE WHEN p.format = 'Test' THEN p.runs ELSE 0 END) AS test_runs,
        MAX(CASE WHEN p.format = 'ODI'  THEN p.runs ELSE 0 END) AS odi_runs,
        MAX(CASE WHEN p.format = 'T20I' THEN p.runs ELSE 0 END) AS t20i_runs,
        ROUND(AVG(p.batting_avg), 2) AS overall_avg
    FROM player_format_performance p
    INNER JOIN (
        SELECT player_name
        FROM player_format_performance
        GROUP BY player_name
        HAVING COUNT(DISTINCT format) >= 2
    ) sub ON p.player_name = sub.player_name
    GROUP BY p.player_name
    ORDER BY overall_avg DESC
""")
rows = cursor.fetchall()
res = pd.DataFrame(rows, columns=["Player", "Test Runs", "ODI Runs", "T20I Runs", "Overall Avg"])
print(res.to_string(index=False))

conn.close()
```

```python


import pandas as pd
import mysql.connector

# --- SAMPLE DATA ---
data = [
    # [team1, team2, winner, venue_country]
    ["India",       "Australia",    "India",        "India"],        # India HOME win
    ["India",       "England",      "India",        "India"],        # India HOME win
    ["India",       "Sri Lanka",    "India",        "India"],        # India HOME win
    ["India",       "New Zealand",  "India",        "India"],        # India HOME win
    ["Australia",   "India",        "India",        "Australia"],    # India AWAY win
    ["England",     "India",        "India",        "England"],      # India AWAY win
    ["West Indies", "India",        "India",        "West Indies"],  # India AWAY win

    ["Australia",   "Pakistan",     "Australia",    "Australia"],    # Australia HOME win
    ["Australia",   "South Africa", "Australia",    "Australia"],    # Australia HOME win
    ["Australia",   "West Indies",  "Australia",    "Australia"],    # Australia HOME win
    ["India",       "Australia",    "Australia",    "India"],        # Australia AWAY win
    ["England",     "Australia",    "Australia",    "England"],      # Australia AWAY win

    ["England",     "Pakistan",     "England",      "England"],      # England HOME win
    ["England",     "New Zealand",  "England",      "England"],      # England HOME win
    ["Australia",   "England",      "England",      "Australia"],    # England AWAY win
    ["India",       "England",      "England",      "India"],        # England AWAY win (upset)

    ["Pakistan",    "Bangladesh",   "Pakistan",     "Pakistan"],     # Pakistan HOME win
    ["Pakistan",    "Australia",    "Pakistan",     "Pakistan"],     # Pakistan HOME win
    ["India",       "Pakistan",     "Pakistan",     "India"],        # Pakistan AWAY win

    ["South Africa","New Zealand",  "South Africa", "South Africa"], # SA HOME win
    ["South Africa","England",      "South Africa", "South Africa"], # SA HOME win
    ["Australia",   "South Africa", "South Africa", "Australia"],    # SA AWAY win

    ["New Zealand", "India",        "New Zealand",  "New Zealand"],  # NZ HOME win
    ["England",     "New Zealand",  "New Zealand",  "England"],      # NZ AWAY win
]

columns = ["team1", "team2", "winner", "venue_country"]
df = pd.DataFrame(data, columns=columns)

# --- PANDAS PROCESSING ---
# For each match, determine if the winner played at home or away
records = []
for _, row in df.iterrows():
    winner = row["winner"]
    venue_country = row["venue_country"]

    # Determine winner's country (same as team name in this dataset)
    if winner == venue_country:
        records.append({"team": winner, "condition": "Home"})
    else:
        records.append({"team": winner, "condition": "Away"})

wins_df = pd.DataFrame(records)
pivot = wins_df.groupby(["team", "condition"]).size().unstack(fill_value=0).reset_index()
if "Home" not in pivot.columns:
    pivot["Home"] = 0
if "Away" not in pivot.columns:
    pivot["Away"] = 0
pivot = pivot[["team", "Home", "Away"]]
pivot["Total Wins"] = pivot["Home"] + pivot["Away"]
pivot = pivot.sort_values("Total Wins", ascending=False).reset_index(drop=True)
pivot.index += 1
pivot.index.name = "Rank"

print("=" * 55)
print("QUESTION 12: Home vs Away Performance")
print("=" * 55)
print(pivot.rename(columns={"team": "Team", "Home": "Home Wins", "Away": "Away Wins"}).to_string())

# --- MySQL ---
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Elan@2004",
    database="cricbuzz_livestats"
)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS home_away_matches (
    id INT AUTO_INCREMENT PRIMARY KEY,
    team1 VARCHAR(50),
    team2 VARCHAR(50),
    winner VARCHAR(50),
    venue_country VARCHAR(50)
)
""")
cursor.execute("DELETE FROM home_away_matches")

for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO home_away_matches (team1, team2, winner, venue_country)
        VALUES (%s, %s, %s, %s)
    """, (row["team1"], row["team2"], row["winner"], row["venue_country"]))

conn.commit()

# --- SQL QUERY ---
print("\n--- SQL QUERY OUTPUT ---")
cursor.execute("""
    SELECT 
        winner AS team,
        SUM(CASE WHEN winner = venue_country THEN 1 ELSE 0 END) AS home_wins,
        SUM(CASE WHEN winner != venue_country THEN 1 ELSE 0 END) AS away_wins,
        COUNT(*) AS total_wins
    FROM home_away_matches
    GROUP BY winner
    ORDER BY total_wins DESC
""")
rows = cursor.fetchall()
res = pd.DataFrame(rows, columns=["Team", "Home Wins", "Away Wins", "Total Wins"])
print(res.to_string(index=False))

conn.close()
```

```python


import pandas as pd
import mysql.connector

# --- SAMPLE DATA ---
data = [
    # [player, team, match_id, runs_scored, match_margin_value, margin_type, team_won]
    ["Virat Kohli",     "India",     1, 72, 32,  "runs",    True],   # Close (32 < 50) 
    ["Virat Kohli",     "India",     2, 55, 3,   "wickets", True],   # Close (3 < 5) 
    ["Virat Kohli",     "India",     3, 40, 80,  "runs",    False],  # NOT close (80 >= 50) 
    ["Virat Kohli",     "India",     4, 88, 2,   "wickets", True],   # Close 
    ["Virat Kohli",     "India",     5, 30, 4,   "wickets", False],  # Close  but team lost

    ["Rohit Sharma",    "India",     1, 60, 32,  "runs",    True],   # Close 
    ["Rohit Sharma",    "India",     2, 45, 3,   "wickets", True],   # Close 
    ["Rohit Sharma",    "India",     3, 78, 80,  "runs",    False],  # NOT close 
    ["Rohit Sharma",    "India",     4, 35, 2,   "wickets", True],   # Close 

    ["Steve Smith",     "Australia", 6, 95, 10,  "runs",    True],   # Close 
    ["Steve Smith",     "Australia", 7, 42, 4,   "wickets", True],   # Close 
    ["Steve Smith",     "Australia", 8, 68, 55,  "runs",    True],   # NOT close 
    ["Steve Smith",     "Australia", 9, 50, 1,   "wickets", False],  # Close  team lost

    ["Babar Azam",      "Pakistan",  10, 80, 22, "runs",    True],   # Close 
    ["Babar Azam",      "Pakistan",  11, 55, 3,  "wickets", True],   # Close 
    ["Babar Azam",      "Pakistan",  12, 30, 4,  "wickets", False],  # Close  team lost
    ["Babar Azam",      "Pakistan",  13, 45, 60, "runs",    True],   # NOT close 

    ["Joe Root",        "England",   14, 65, 15, "runs",    True],   # Close 
    ["Joe Root",        "England",   15, 50, 2,  "wickets", False],  # Close  team lost
    ["Joe Root",        "England",   16, 88, 3,  "wickets", True],   # Close 
]

columns = ["player", "team", "match_id", "runs_scored", "margin_value", "margin_type", "team_won"]
df = pd.DataFrame(data, columns=columns)

# --- PANDAS PROCESSING ---
# Identify close matches
df["is_close"] = ((df["margin_type"] == "runs") & (df["margin_value"] < 50)) | \
                 ((df["margin_type"] == "wickets") & (df["margin_value"] < 5))

close_matches = df[df["is_close"]]

result = close_matches.groupby("player").agg(
    avg_runs=("runs_scored", "mean"),
    total_close_matches=("match_id", "count"),
    team_wins=("team_won", "sum")
).reset_index()

result["avg_runs"] = result["avg_runs"].round(2)
result["team_wins"] = result["team_wins"].astype(int)
result = result.sort_values("avg_runs", ascending=False).reset_index(drop=True)
result.index += 1
result.index.name = "Rank"

print("=" * 75)
print("QUESTION 15: Players' Performance in Close Matches")
print("=" * 75)
print(result.rename(columns={
    "player": "Player", "avg_runs": "Avg Runs in Close Matches",
    "total_close_matches": "Close Matches Played", "team_wins": "Team Wins"
}).to_string())

# --- MySQL ---
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Elan@2004",
    database="cricbuzz_livestats"
)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS close_match_performance (
    id INT AUTO_INCREMENT PRIMARY KEY,
    player VARCHAR(100),
    team VARCHAR(50),
    match_id INT,
    runs_scored INT,
    margin_value INT,
    margin_type VARCHAR(10),
    team_won BOOLEAN
)
""")
cursor.execute("DELETE FROM close_match_performance")

for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO close_match_performance
        (player, team, match_id, runs_scored, margin_value, margin_type, team_won)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
    """, (row["player"], row["team"], int(row["match_id"]), int(row["runs_scored"]),
          int(row["margin_value"]), row["margin_type"], bool(row["team_won"])))

conn.commit()

# --- SQL QUERY ---
print("\n--- SQL QUERY OUTPUT ---")
cursor.execute("""
    SELECT player,
           ROUND(AVG(runs_scored), 2) AS avg_runs,
           COUNT(*) AS total_close_matches,
           SUM(CASE WHEN team_won = TRUE THEN 1 ELSE 0 END) AS team_wins
    FROM close_match_performance
    WHERE (margin_type = 'runs' AND margin_value < 50)
       OR (margin_type = 'wickets' AND margin_value < 5)
    GROUP BY player
    ORDER BY avg_runs DESC
""")
rows = cursor.fetchall()
res = pd.DataFrame(rows, columns=["Player", "Avg Runs", "Close Matches", "Team Wins"])
print(res.to_string(index=False))

conn.close()
```

```python
#25

import pandas as pd
import mysql.connector
import numpy as np
from datetime import datetime

# --- SAMPLE DATA (structured quarterly, NOT random) ---
# Each row: [player, quarter, match_count, avg_runs, avg_strike_rate]
data = [
    # Virat Kohli - Career Ascending (steady improvement)
    ["Virat Kohli", "2020Q1", 4, 42.0, 128.0],
    ["Virat Kohli", "2020Q2", 4, 45.0, 130.0],
    ["Virat Kohli", "2020Q3", 3, 48.0, 132.0],
    ["Virat Kohli", "2020Q4", 4, 50.0, 135.0],
    ["Virat Kohli", "2021Q1", 3, 52.0, 138.0],
    ["Virat Kohli", "2021Q2", 4, 55.0, 140.0],
    ["Virat Kohli", "2021Q3", 3, 58.0, 142.0],
    ["Virat Kohli", "2021Q4", 4, 60.0, 145.0],

    # Rohit Sharma - Career Stable (fluctuates around same level)
    ["Rohit Sharma", "2020Q1", 4, 52.0, 135.0],
    ["Rohit Sharma", "2020Q2", 3, 48.0, 130.0],
    ["Rohit Sharma", "2020Q3", 4, 55.0, 138.0],
    ["Rohit Sharma", "2020Q4", 3, 50.0, 132.0],
    ["Rohit Sharma", "2021Q1", 4, 53.0, 136.0],
    ["Rohit Sharma", "2021Q2", 3, 47.0, 129.0],
    ["Rohit Sharma", "2021Q3", 4, 54.0, 137.0],
    ["Rohit Sharma", "2021Q4", 3, 51.0, 133.0],

    # Steve Smith - Career Declining (drops over time)
    ["Steve Smith", "2020Q1", 4, 75.0, 82.0],
    ["Steve Smith", "2020Q2", 3, 70.0, 78.0],
    ["Steve Smith", "2020Q3", 4, 62.0, 72.0],
    ["Steve Smith", "2020Q4", 3, 55.0, 68.0],
    ["Steve Smith", "2021Q1", 4, 48.0, 64.0],
    ["Steve Smith", "2021Q2", 3, 40.0, 58.0],
    ["Steve Smith", "2021Q3", 4, 35.0, 52.0],
    ["Steve Smith", "2021Q4", 3, 28.0, 48.0],

    # Babar Azam - Career Ascending
    ["Babar Azam", "2020Q1", 3, 38.0, 125.0],
    ["Babar Azam", "2020Q2", 4, 42.0, 128.0],
    ["Babar Azam", "2020Q3", 3, 48.0, 132.0],
    ["Babar Azam", "2020Q4", 4, 52.0, 136.0],
    ["Babar Azam", "2021Q1", 3, 56.0, 140.0],
    ["Babar Azam", "2021Q2", 4, 60.0, 143.0],
    ["Babar Azam", "2021Q3", 3, 62.0, 145.0],
    ["Babar Azam", "2021Q4", 4, 65.0, 148.0],

    # Joe Root - Career Stable
    ["Joe Root", "2020Q1", 4, 58.0, 72.0],
    ["Joe Root", "2020Q2", 3, 55.0, 70.0],
    ["Joe Root", "2020Q3", 4, 60.0, 74.0],
    ["Joe Root", "2020Q4", 3, 56.0, 71.0],
    ["Joe Root", "2021Q1", 4, 62.0, 75.0],
    ["Joe Root", "2021Q2", 3, 57.0, 72.0],
    ["Joe Root", "2021Q3", 4, 59.0, 73.0],
    ["Joe Root", "2021Q4", 3, 61.0, 74.0],

    # David Warner - Career Declining
    ["David Warner", "2020Q1", 4, 68.0, 140.0],
    ["David Warner", "2020Q2", 3, 60.0, 132.0],
    ["David Warner", "2020Q3", 4, 52.0, 125.0],
    ["David Warner", "2020Q4", 3, 45.0, 118.0],
    ["David Warner", "2021Q1", 4, 38.0, 110.0],
    ["David Warner", "2021Q2", 3, 32.0, 105.0],
    ["David Warner", "2021Q3", 4, 28.0, 98.0],
    ["David Warner", "2021Q4", 3, 22.0, 92.0],

    # Only 4 quarters - should be EXCLUDED (< 6 quarters)
    ["Pat Cummins", "2020Q1", 3, 12.0, 65.0],
    ["Pat Cummins", "2020Q2", 3, 15.0, 70.0],
    ["Pat Cummins", "2020Q3", 3, 10.0, 62.0],
    ["Pat Cummins", "2020Q4", 3, 14.0, 68.0],
]

columns = ["player", "quarter", "matches", "avg_runs", "avg_sr"]
df = pd.DataFrame(data, columns=columns)

# --- PANDAS PROCESSING ---
# Filter: >= 3 matches per quarter (already satisfied in our data)
quarterly = df[df["matches"] >= 3].copy()

# Count quarters per player
quarter_counts = quarterly.groupby("player")["quarter"].count()
valid_players = quarter_counts[quarter_counts >= 6].index.tolist()

# Filter to valid players only
valid_df = quarterly[quarterly["player"].isin(valid_players)].copy()

# Calculate quarter-to-quarter trends
def get_trend(curr, prev):
    if prev == 0:
        return "Stable"
    change = (curr - prev) / abs(prev) * 100
    if change > 5:
        return "Improving"
    elif change < -5:
        return "Declining"
    else:
        return "Stable"

trend_records = []
for player in valid_df["player"].unique():
    player_data = valid_df[valid_df["player"] == player].sort_values("quarter")
    prev_runs, prev_sr = None, None

    for _, row in player_data.iterrows():
        if prev_runs is not None:
            trend_runs = get_trend(row["avg_runs"], prev_runs)
            trend_sr = get_trend(row["avg_sr"], prev_sr)
        else:
            trend_runs = "Stable"
            trend_sr = "Stable"

        trend_records.append({
            "player": row["player"],
            "quarter": row["quarter"],
            "avg_runs": round(row["avg_runs"], 2),
            "avg_sr": round(row["avg_sr"], 2),
            "runs_trend": trend_runs,
            "sr_trend": trend_sr
        })
        prev_runs = row["avg_runs"]
        prev_sr = row["avg_sr"]

trend_df = pd.DataFrame(trend_records)

# --- CAREER PHASE CATEGORIZATION ---
def career_phase(sub_df):
    improving = ((sub_df["runs_trend"] == "Improving") | (sub_df["sr_trend"] == "Improving")).sum()
    declining = ((sub_df["runs_trend"] == "Declining") | (sub_df["sr_trend"] == "Declining")).sum()

    if improving > declining * 1.5:
        return "Career Ascending"
    elif declining > improving * 1.5:
        return "Career Declining"
    else:
        return "Career Stable"

career_df = trend_df.groupby("player").apply(career_phase).reset_index(name="career_phase")

# Get latest quarter per player
latest_df = trend_df.sort_values("quarter").groupby("player").tail(1).copy()
final_df = pd.merge(latest_df, career_df, on="player")

# Also get first quarter for comparison
first_df = trend_df.sort_values("quarter").groupby("player").head(1)[["player", "avg_runs", "avg_sr"]].rename(
    columns={"avg_runs": "first_avg_runs", "avg_sr": "first_avg_sr"}
)
final_df = pd.merge(final_df, first_df, on="player")
final_df["runs_change"] = (final_df["avg_runs"] - final_df["first_avg_runs"]).round(2)

final_df = final_df.sort_values("career_phase").reset_index(drop=True)
final_df.index += 1
final_df.index.name = "Sl.No"

print("=" * 100)
print("QUESTION 25: Player Performance Evolution (Career Phase Analysis)")
print("=" * 100)
print("\n--- Full Quarterly Trend Data ---")
print(trend_df.to_string(index=False))

print("\n--- Career Phase Summary (Latest Quarter) ---")
print(final_df[["player", "quarter", "avg_runs", "avg_sr", "runs_trend", "sr_trend",
                "first_avg_runs", "runs_change", "career_phase"]].rename(columns={
    "player": "Player", "quarter": "Latest Quarter",
    "avg_runs": "Latest Avg Runs", "avg_sr": "Latest Avg SR",
    "runs_trend": "Runs Trend", "sr_trend": "SR Trend",
    "first_avg_runs": "Starting Avg Runs", "runs_change": "Total Change",
    "career_phase": "Career Phase"
}).to_string())

# --- MySQL ---
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Elan@2004",
    database="cricbuzz_livestats"
)
cursor = conn.cursor()

# Quarterly trends table (all quarters)
cursor.execute("""
CREATE TABLE IF NOT EXISTS player_quarterly_trends (
    id INT AUTO_INCREMENT PRIMARY KEY,
    player VARCHAR(100),
    quarter VARCHAR(10),
    avg_runs FLOAT,
    avg_sr FLOAT,
    runs_trend VARCHAR(20),
    sr_trend VARCHAR(20)
)
""")
cursor.execute("DELETE FROM player_quarterly_trends")

for _, row in trend_df.iterrows():
    cursor.execute("""
        INSERT INTO player_quarterly_trends (player, quarter, avg_runs, avg_sr, runs_trend, sr_trend)
        VALUES (%s, %s, %s, %s, %s, %s)
    """, (row["player"], row["quarter"], float(row["avg_runs"]),
          float(row["avg_sr"]), row["runs_trend"], row["sr_trend"]))

# Career phase summary table
cursor.execute("""
CREATE TABLE IF NOT EXISTS player_performance_trends (
    id INT AUTO_INCREMENT PRIMARY KEY,
    player VARCHAR(100),
    quarter VARCHAR(10),
    avg_runs FLOAT,
    avg_sr FLOAT,
    runs_trend VARCHAR(20),
    sr_trend VARCHAR(20),
    career_phase VARCHAR(30)
)
""")
cursor.execute("DELETE FROM player_performance_trends")

for _, row in final_df.iterrows():
    cursor.execute("""
        INSERT INTO player_performance_trends
        (player, quarter, avg_runs, avg_sr, runs_trend, sr_trend, career_phase)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
    """, (row["player"], row["quarter"], float(row["avg_runs"]),
          float(row["avg_sr"]), row["runs_trend"], row["sr_trend"], row["career_phase"]))

conn.commit()

# --- SQL QUERY ---
print("\n--- SQL QUERY OUTPUT (Career Phase Summary) ---")
cursor.execute("""
    SELECT player, quarter, avg_runs, avg_sr, runs_trend, sr_trend, career_phase
    FROM player_performance_trends
    ORDER BY career_phase, player
""")
rows = cursor.fetchall()
res = pd.DataFrame(rows, columns=["Player", "Quarter", "Avg Runs", "Avg SR",
                                   "Runs Trend", "SR Trend", "Career Phase"])
print(res.to_string(index=False))

print("\n--- SQL QUERY OUTPUT (All Quarterly Data) ---")
cursor.execute("""
    SELECT player, quarter, avg_runs, avg_sr, runs_trend, sr_trend
    FROM player_quarterly_trends
    ORDER BY player, quarter
""")
rows = cursor.fetchall()
res2 = pd.DataFrame(rows, columns=["Player", "Quarter", "Avg Runs", "Avg SR", "Runs Trend", "SR Trend"])
print(res2.to_string(index=False))

conn.close()
```
